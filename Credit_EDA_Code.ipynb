## Imputaion for missing Values



1.   AMT_ANNUITY cannot be derived from any other column in the dataset.
While we examine the 12 records that are missing , we see that the target column is 0 for all the 12 values of missing AMT_ANNUITY. 
Thus **best option for imputing the missing values** is to** drop the records that have missing values** in this particular column as they **dont contribute more to the target variable imbalance**.
2.   AMT_GOODS_PRICE has 278 which is only 0.09% of the data,either mean / median imputation can be used since the data is highly skewed. 
3.   CNT_FAM_MEMBERS has 2 records with missing value, since its a categorical value, one way of imputing such a **small percentage of missing value is to replace it with the MODE value**.
4.   OCCUPATION_TYPE has a 31.34% of missing values , but may be a very important factor in the EDA , so the records cannot be dropped. A suitable method for handling these records would be to **rename them to a new category "Unknown"**. This helps us to perserve the variance of the variable.
5.   NAME_TYPE_SUITE since its a categorical value with less significant percentage of missing values, the column can be imputed using MODE value.

## Data Imbalance

The distribution of records with Target==0 , makes up 92 % of the records in the data set , on the other hand , for Target==1 it only constitutes to 8% of the datset records.
####Methods to fix data imbalance

*   Oversampling can be used when the quantity of data is insufficient. It tries to balance dataset by increasing the size of fewer samples(Target==1). Rather than getting rid of abundant samples, new rare samples are generated by using e.g. repetition, bootstrapping.
*   Under-sampling balances the dataset by reducing the size of the abundant class. This method is used when quantity of data is sufficient. By keeping all samples in the rare class and randomly selecting an equal number of samples in the abundant class, a balanced new dataset can be retrieved.
